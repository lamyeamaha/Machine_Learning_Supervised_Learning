```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/datasets/Linear Regression_canada_per_capita_income/canada_per_capita_income.csv")
df.head()
df.columns
```

```python
# plot scatter plot
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df[['year']], df[['per capita income (US$)']], c='red', marker='+')
plt.plot(df[['year']], df[['per capita income (US$)']], c='blue')
```

```python
reg = linear_model.LinearRegression() #object
reg.fit(df[['year']], df[['per capita income (US$)']])
```

```python
#get the trained coefficient and intercept
print("Optimum coefficient: ",reg.coef_)
print("Optimum intercept: ", reg.intercept_)
```

```python
reg.predict([[2020]])
```

```python
#plot the line:
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df[['year']], df[['per capita income (US$)']], c='red', marker='+')
plt.plot(df[['year']], reg.predict(df[['year']]), c='blue')
```
