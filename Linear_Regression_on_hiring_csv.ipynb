# **Import11**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# **Read data**

```python
df = pd.read_csv("/content/drive/MyDrive/datasets/Linear Regression/hiring.csv")
df.head()
```

# **Data Cleaning**

```python
print("Number of null values in experience column: ", df['experience'].isnull().values.sum())
print("Number of null values in test_score column: ", df['test_score(out of 10)'].isnull().values.sum())
print("Number of null values in interview_score column: ", df['interview_score(out of 10)'].isnull().values.sum())
```

```python
#fill null values in experience column with 0
df['experience'] = df['experience'].fillna("zero")
print("Number of null values in experience column: ", df['experience'].isnull().values.sum())
df
```

```python
#convert words in experience column with numbers:
!pip install word2number
from word2number import w2n
```

```python
type(df['experience'])
```

```python
def convertToNum(word):
  return w2n.word_to_num(word)


df['experience']=df['experience'].apply(convertToNum)
df.head()
```

```python
#fill in missing values of test score
mean_value = df['test_score(out of 10)'].mean()
mean_value
```

```python

df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(mean_value)
df
```

# **Linear Regression Model**

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df[['salary($)']])
```

```python
#check the coefficients and intercepts:
print("coefficients: ",reg.coef_)
print("intercept:", reg.intercept_)
```

# **Make Predictions**

```python
#predict:
reg.predict([[2,9,6]])
```
